# Training the model

```python
# importing necessary libraries
import os
from ultralytics import YOLO
import torch
from IPython.display import Image, display
```

```python
# verify CUDA availability (GPU)
print(f"CUDA available: {torch.cuda.is_available()}") # gpu nh h :(
```

```python
# initialize YOLO model
model = YOLO('yolo11n.pt')
```

```python
# train the model
results = model.train(
    data='dataset/yolo/data.yaml',
    epochs=25,
    imgsz=640,
    batch=8,
    device='cuda' if torch.cuda.is_available() else 'cpu',
    name='parking_cars_yolo12',
    optimizer='auto',
    lr0=0.01,
    cos_lr=True,
    augment=True,
    patience=30,
    seed=42
)
```

```python
# evaluate on test set
metrics = model.val(split='test')
```

```python
# run inference
results = model.predict(
    source='dataset/yolo/test/images',
    conf=0.5,
    save=True,
    show_labels=True,
    show_conf=True
)
```

# Testing the model

```python
model = YOLO("runs/detect/parking_cars_yolo12/weights/best.pt")
model.export(format='onnx')
```

```python
model.save('models/best_model.pt')
```

```python
test_images_dir = "dataset/yolo/test/images"
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def test_and_visualize(model, test_dir, num_samples=3):
    test_images = [os.path.join(test_dir, f) for f in os.listdir(test_dir)]
    
    results = model.predict(
        source=test_images,
        conf=0.5,
        save=True,
        save_txt=True,
        save_conf=True,
        exist_ok=True
    )
    
    plt.figure(figsize=(15, 10))
    for i, (img_path, result) in enumerate(zip(test_images[:num_samples], results[:num_samples])):
        plt.subplot(num_samples, 2, 2*i+1)
        plt.imshow(plt.imread(img_path))
        plt.title('Original Image')
        plt.axis('off')
        
        plt.subplot(num_samples, 2, 2*i+2)
        plotted = result.plot()  # Returns RGB numpy array
        plt.imshow(plotted)
        plt.title('Prediction')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
test_and_visualize(model, test_images_dir, num_samples=8)
```
