```python
import cv2
import tensorflow as tf
import numpy as np
from ultralytics import YOLO
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
unet_model = tf.keras.models.load_model('best_model.keras')
yolo_model = YOLO("best_model.pt")
```

```python
# 2. Define color scheme
COLORS = {
    'free': (0, 255, 0),    # Green
    'occupied': (0, 0, 255), # Red
    'text': (255, 255, 255)  # White
}
```

```python
# 3. Parking space analysis function
def analyze_parking_space(image_path):
    # Read and preprocess image
    original_img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    img_height, img_width = original_img.shape[:2]
    
    # UNet prediction
    unet_input = cv2.resize(original_img, (256, 256)) / 255.0
    mask = unet_model.predict(np.expand_dims(unet_input, axis=0))[0]
    parking_mask = (mask.squeeze() > 0.5).astype(np.uint8)
    parking_mask = cv2.resize(parking_mask, (img_width, img_height))
    
    # Find parking space contours
    contours, _ = cv2.findContours(parking_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # YOLO prediction
    yolo_results = yolo_model.predict(original_img, conf=0.5)
    car_boxes = yolo_results[0].boxes.xyxy.cpu().numpy()
    
    # Create visualization image
    overlay = original_img.copy()
    occupied_count = 0
    
    # Analyze each parking space
    for contour in contours:
        # Get parking space bounding rect
        x, y, w, h = cv2.boundingRect(contour)
        
        # Check for car presence
        space_occupied = False
        for car_box in car_boxes:
            car_x1, car_y1, car_x2, car_y2 = car_box
            # Check overlap between car and parking space
            if (x < car_x1 < x + w and y < car_y1 < y + h) or \
               (x < car_x2 < x + w and y < car_y2 < y + h):
                space_occupied = True
                break
        
        # Update count and draw rectangle
        if space_occupied:
            occupied_count += 1
            color = COLORS['occupied']
        else:
            color = COLORS['free']
            
        cv2.rectangle(overlay, (x, y), (x+w, y+h), color, 2)
    
    # Add count text
    total_spaces = len(contours)
    free_count = total_spaces - occupied_count
    text = f"Occupied: {occupied_count} | Free: {free_count}"
    cv2.putText(overlay, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 
               1, COLORS['text'], 2, cv2.LINE_AA)
    
    return original_img, overlay, occupied_count
```

```python
# 4. Visualization function
def plot_results(original, overlay):
    plt.figure(figsize=(15, 8))
    
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(overlay)
    plt.title('Parking Space Analysis')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# 5. Test integration on sample image
test_image_path = 'data/test/(8).jpg'  # Replace with your image
original, result, count = analyze_parking_space(test_image_path)
plot_results(original, result)
```

```python
# 2. Define color scheme
COLORS = {
    'free': (0, 255, 0),    # Green
    'occupied': (0, 0, 255), # Red
    'text': (255, 255, 255)  # White
}

# 3. Improved parking space analysis function
def analyze_parking_space(image_path):
    # Read and preprocess image
    original_img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    img_height, img_width = original_img.shape[:2]
    
    # UNet prediction
    unet_input = cv2.resize(original_img, (256, 256)) / 255.0
    mask = unet_model.predict(np.expand_dims(unet_input, axis=0))[0]
    parking_mask = (mask.squeeze() > 0.5).astype(np.uint8)
    parking_mask = cv2.resize(parking_mask, (img_width, img_height))
    
    # Find parking space contours
    contours, _ = cv2.findContours(parking_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # YOLO prediction
    yolo_results = yolo_model.predict(original_img, conf=0.5)
    car_boxes = yolo_results[0].boxes.xyxy.cpu().numpy()
    
    # Create visualization image
    overlay = original_img.copy()
    occupied_count = 0
    
    # Analyze each parking space
    for contour in contours:
        # Get parking space bounding rect
        x, y, w, h = cv2.boundingRect(contour)
        
        # Check for car presence using Intersection over Union (IoU)
        space_occupied = False
        for car_box in car_boxes:
            car_x1, car_y1, car_x2, car_y2 = car_box
            # Calculate IoU between parking space and car bounding box
            intersection_area = max(0, min(x + w, car_x2) - max(x, car_x1)) * max(0, min(y + h, car_y2) - max(y, car_y1))
            parking_area = w * h
            car_area = (car_x2 - car_x1) * (car_y2 - car_y1)
            iou = intersection_area / (parking_area + car_area - intersection_area)
            
            # If IoU > threshold, mark as occupied
            if iou > 0.2:  # Adjust threshold as needed
                space_occupied = True
                break
        
        # Update count and draw rectangle
        if space_occupied:
            occupied_count += 1
            color = COLORS['occupied']
        else:
            color = COLORS['free']
            
        cv2.rectangle(overlay, (x, y), (x + w, y + h), color, 2)
    
    # Add count text
    total_spaces = len(contours)
    free_count = total_spaces - occupied_count
    text = f"Occupied: {occupied_count} | Free: {free_count}"
    cv2.putText(overlay, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 
               1, COLORS['text'], 2, cv2.LINE_AA)
    
    return original_img, overlay, occupied_count

# 4. Visualization function
def plot_results(original, overlay):
    plt.figure(figsize=(15, 8))
    
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(overlay)
    plt.title('Parking Space Analysis')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# 5. Test integration on sample image
test_image_path = 'data/test/(8).jpg'  # Replace with your image
original, result, count = analyze_parking_space(test_image_path)
plot_results(original, result)
```

```python
# Color scheme
COLORS = {
    'free': (0, 255, 0),
    'occupied': (0, 0, 255),
    'text': (255, 255, 255)
}

def merge_contours(contours, max_distance=50):
    """Merge nearby contours into single parking spaces"""
    merged_rects = []
    
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        
        # Check if this rectangle can be merged with existing ones
        merged = False
        for i, (mx, my, mw, mh) in enumerate(merged_rects):
            # Calculate distance between rectangle centers
            cx = x + w/2
            cy = y + h/2
            mcx = mx + mw/2
            mcy = my + mh/2
            distance = np.sqrt((cx - mcx)**2 + (cy - mcy)**2)
            
            if distance < max_distance:
                # Merge rectangles
                new_x = min(x, mx)
                new_y = min(y, my)
                new_w = max(x+w, mx+mw) - new_x
                new_h = max(y+h, my+mh) - new_y
                merged_rects[i] = (new_x, new_y, new_w, new_h)
                merged = True
                break
                
        if not merged:
            merged_rects.append((x, y, w, h))
            
    return merged_rects

def analyze_parking_space(image_path):
    # Read and preprocess image
    original_img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    img_height, img_width = original_img.shape[:2]
    
    # UNet prediction with post-processing
    unet_input = cv2.resize(original_img, (256, 256)) / 255.0
    mask = unet_model.predict(np.expand_dims(unet_input, axis=0))[0]
    parking_mask = (mask.squeeze() > 0.5).astype(np.uint8)
    
    # Post-processing steps
    parking_mask = cv2.resize(parking_mask, (img_width, img_height))
    
    # 1. Morphological closing to merge nearby regions
    kernel = np.ones((15, 15), np.uint8)
    parking_mask = cv2.morphologyEx(parking_mask, cv2.MORPH_CLOSE, kernel)
    
    # 2. Remove small noise
    parking_mask = cv2.morphologyEx(parking_mask, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8))
    
    # Find and merge contours
    contours, _ = cv2.findContours(parking_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 3. Filter small contours
    min_area = 500  # Adjust based on your parking space size
    filtered_contours = [c for c in contours if cv2.contourArea(c) > min_area]
    
    # 4. Merge nearby contours
    merged_rects = merge_contours(filtered_contours)
    
    # YOLO prediction
    yolo_results = yolo_model.predict(original_img, conf=0.5)
    car_boxes = yolo_results[0].boxes.xyxy.cpu().numpy()
    
    # Create visualization image
    overlay = original_img.copy()
    occupied_count = 0
    
    # Analyze each merged parking space
    for (x, y, w, h) in merged_rects:
        # Check center point occupancy
        space_center = (x + w//2, y + h//2)
        occupied = False
        
        for car_box in car_boxes:
            car_x1, car_y1, car_x2, car_y2 = car_box
            if (car_x1 < space_center[0] < car_x2 and 
                car_y1 < space_center[1] < car_y2):
                occupied = True
                break
                
        # Update counts and visualization
        if occupied:
            occupied_count += 1
            color = COLORS['occupied']
        else:
            color = COLORS['free']
            
        cv2.rectangle(overlay, (x, y), (x+w, y+h), color, 2)
    
    # Add count text
    total_spaces = len(merged_rects)
    free_count = total_spaces - occupied_count
    text = f"Occupied: {occupied_count} | Free: {free_count}"
    cv2.putText(overlay, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 
               1, COLORS['text'], 2, cv2.LINE_AA)
    
    return original_img, overlay, occupied_count


# 4. Visualization function
def plot_results(original, overlay):
    plt.figure(figsize=(15, 8))
    
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(overlay)
    plt.title('Parking Space Analysis')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# 5. Test integration on sample image
test_image_path = 'data/test/(8).jpg'  # Replace with your image
original, result, count = analyze_parking_space(test_image_path)
plot_results(original, result)
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
%matplotlib inline
import matplotlib.pyplot as plt

# Load models
unet_model = tf.keras.models.load_model('best_model.keras')
yolo_model = YOLO("best_model.pt")

# Color scheme
COLORS = {
    'free': (0, 255, 0),    # Green
    'occupied': (0, 0, 255), # Red
    'text': (255, 255, 255)  # White
}

def calculate_iou(box1, box2):
    """Calculate Intersection over Union between two rectangles"""
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])
    
    intersection = max(0, x2 - x1) * max(0, y2 - y1)
    area1 = (box1[2]-box1[0]) * (box1[3]-box1[1])
    area2 = (box2[2]-box2[0]) * (box2[3]-box2[1])
    
    return intersection / (area1 + area2 - intersection + 1e-6)

def merge_contours(contours, max_distance=75):
    """Merge nearby contours with improved logic"""
    merged = []
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        found = False
        for idx, (mx, my, mw, mh) in enumerate(merged):
            if (abs(x - mx) < max_distance and 
                abs(y - my) < max_distance and 
                abs(x+w - mx-mw) < max_distance):
                # Merge rectangles
                new_x = min(x, mx)
                new_y = min(y, my)
                new_w = max(x+w, mx+mw) - new_x
                new_h = max(y+h, my+mh) - new_y
                merged[idx] = (new_x, new_y, new_w, new_h)
                found = True
                break
        if not found:
            merged.append((x, y, w, h))
    return merged

def analyze_parking_space(image_path):
    # Load and process image
    original_img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    h, w = original_img.shape[:2]
    
    # UNet processing with enhanced morphology
    unet_input = cv2.resize(original_img, (256, 256)) / 255.0
    mask = unet_model.predict(np.expand_dims(unet_input, axis=0))[0]
    parking_mask = (mask.squeeze() > 0.5).astype(np.uint8)
    parking_mask = cv2.resize(parking_mask, (w, h))
    
    # Morphological operations
    kernel = np.ones((25,25), np.uint8)
    parking_mask = cv2.morphologyEx(parking_mask, cv2.MORPH_CLOSE, kernel)
    parking_mask = cv2.morphologyEx(parking_mask, cv2.MORPH_OPEN, np.ones((7,7), np.uint8))
    
    # Contour processing
    contours, _ = cv2.findContours(parking_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    filtered = [c for c in contours if cv2.contourArea(c) > 1500]
    merged_rects = merge_contours(filtered)
    
    # YOLO detection
    results = yolo_model.predict(original_img, conf=0.4)
    car_boxes = results[0].boxes.xyxy.cpu().numpy()
    
    # Parking space analysis
    overlay = original_img.copy()
    occupied = 0
    
    for rect in merged_rects:
        x, y, w, h = rect
        space_box = [x, y, x+w, y+h]
        is_occupied = False
        
        for car in car_boxes:
            car_box = [int(c) for c in car]
            iou = calculate_iou(space_box, car_box)
            if iou > 0.2:  # IoU threshold
                is_occupied = True
                break
        
        color = COLORS['occupied'] if is_occupied else COLORS['free']
        cv2.rectangle(overlay, (x, y), (x+w, y+h), color, 2)
        occupied += int(is_occupied)
    
    # Add info text
    total = len(merged_rects)
    text = f"Occupied: {occupied} | Free: {total - occupied}"
    cv2.putText(overlay, text, (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 
               1, COLORS['text'], 2, cv2.LINE_AA)
    
    return original_img, overlay, occupied

def plot_results(original, overlay):
    plt.figure(figsize=(15, 8))
    
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(overlay)
    plt.title('Parking Space Analysis')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()


# Test with your image
test_image_path = 'data/test/images/(8).jpg'
original, result, count = analyze_parking_space(test_image_path)
plot_results(original, result)
print(f"Occupied spaces: {count}")
```
